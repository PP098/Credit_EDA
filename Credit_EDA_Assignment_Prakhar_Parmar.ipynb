**We can also remove the rows if null count is low**

```python
application=application[~application["NAME_TYPE_SUITE"].isnull()]
application=application[~application["OBS_30_CNT_SOCIAL_CIRCLE"].isnull()]
application=application[~application["DEF_30_CNT_SOCIAL_CIRCLE"].isnull()]
application=application[~application["OBS_60_CNT_SOCIAL_CIRCLE"].isnull()]
application=application[~application["DEF_60_CNT_SOCIAL_CIRCLE"].isnull()]
application=application[~application["EXT_SOURCE_2"].isnull()]
application=application[~application["AMT_GOODS_PRICE"].isnull()]
application=application[~application["AMT_ANNUITY"].isnull()]
application=application[~application["CNT_FAM_MEMBERS"].isnull()]
application=application[~application["DAYS_LAST_PHONE_CHANGE"].isnull()]
```

```python
application.shape
```

```python
application.isnull().sum().sort_values(ascending=False).head(10)
```

**We see we still have 8 columns with missing values which need to be treated.**

# In OCCUPATION_TYPE , we can use the mode value to replce the missing values,by getting value counts of it.


Also 
MT_REQ_CREDIT_BUREAU_YEAR   
AMT_REQ_CREDIT_BUREAU_QRT     
AMT_REQ_CREDIT_BUREAU_MON     
AMT_REQ_CREDIT_BUREAU_WEEK    
AMT_REQ_CREDIT_BUREAU_DAY     
AMT_REQ_CREDIT_BUREAU_HOUR 
can be filled with mode value of the column , i.e 0

```python
m=application.OCCUPATION_TYPE.mode()
```

```python
application.OCCUPATION_TYPE=application.OCCUPATION_TYPE.fillna(m[0])
application.AMT_REQ_CREDIT_BUREAU_YEAR=application.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(0)
application.AMT_REQ_CREDIT_BUREAU_QRT=application.AMT_REQ_CREDIT_BUREAU_QRT.fillna(0)
application.AMT_REQ_CREDIT_BUREAU_MON=application.AMT_REQ_CREDIT_BUREAU_MON.fillna(0)
application.AMT_REQ_CREDIT_BUREAU_WEEK=application.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(0)
application.AMT_REQ_CREDIT_BUREAU_DAY=application.AMT_REQ_CREDIT_BUREAU_DAY.fillna(0)
application.AMT_REQ_CREDIT_BUREAU_HOUR=application.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(0)
```

**For the column 'AMT_GOODS_PRICE' & 'AMT_ANNUITY' we can use median to replace the missing values**

```python
application.AMT_GOODS_PRICE=application.AMT_GOODS_PRICE.fillna(application.AMT_GOODS_PRICE.median())
application.AMT_ANNUITY=application.AMT_ANNUITY.fillna(application.AMT_ANNUITY.median())
```

**We check the "CODE_GENDER" column and found there are unwanted values which need to be treated**



```python
application.CODE_GENDER.value_counts()
```

```python
#lets replcae XNA with "F" as its mode
```

```python
application.loc[application.CODE_GENDER == 'XNA','CODE_GENDER'] = 'F'
```
